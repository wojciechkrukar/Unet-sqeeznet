```python
def fileNamesInFolder(myFolder):
    relativeDir = os.path.dirname(os.path.realpath('__file__'))
    #print(relativeDir)
    folderName = os.path.join(relativeDir, myFolder)
    onlyfiles = [f for f in listdir(folderName) if isfile(join(folderName, f))]
    return onlyfiles
```
